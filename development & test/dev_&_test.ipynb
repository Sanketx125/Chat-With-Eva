```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_google_genai import GoogleGenerativeAIEmbeddings, ChatGoogleGenerativeAI
from langchain.vectorstores import chroma
from langchain_core.prompts import PromptTemplate
```

```python
loader = PyPDFLoader("/content/Web Technology.pdf")
pages = loader.load_and_split()


```

```python
pages[17].page_content
```

```python
model = GoogleGenerativeAIEmbeddings(
    model="models/embedding-001",
    temperature=0,
)
```

```python
pip install langchain-chroma
```

```python
vector_store = Chroma(
    embedding_function= model,
    collection_name= 'sample',
    persist_directory= 'db'
)
```

```python
vector_store.add_documents(pages)
```

```python
vector_store.get(include=['metadatas', 'documents'])
```

```python

# Retriever
retriever = vector_store.as_retriever(
    search_type="mmr",
    search_kwargs={"k": 5,  "lambda_mult": 0.5}
)
```

```python
query = 'web development'
result  = retriever.get_relevant_documents(query)
```

```python
retriever.invoke(query)
```

```python
result
```

```python
for i, doc in enumerate(result):
  print(f'\n----------- Result: {i+1} --------------\n')
  print(doc.page_content)
```

```python
prompt = PromptTemplate(
    template="""You are a helpful assistant. Answer ONLY from the provided transcript context.
      If the context is insufficient, just say you don't know.

      {context}
      Question: {question}
    """,
    input_variables=['context', 'question']
)
```

```python
cotext_text ="\n\n".join( doc.page_content for doc in result)
```

```python
final_prompt =prompt.invoke({'context': cotext_text, 'question': query})
```

```python
final_prompt
```

```python
llm =  ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
ai_answer = llm.invoke(final_prompt)
```

```python
ai_answer.content
```

```python
# building chains
```

```python
from langchain_core.runnables import RunnableParallel, RunnablePassthrough, RunnableLambda
from langchain_core.output_parsers import StrOutputParser
```

```python
# format doc for chain
def format_document(result):
  context_text ="\n\n".join( doc.page_content for doc in result)
  return context_text
```

```python
parallel_chain = RunnableParallel(
    context= retriever | RunnableLambda(format_document),
    question= RunnablePassthrough()
)
```

```python
parallel_chain.invoke('HTTP')
```

```python
parser = StrOutputParser()
```

```python
main_chain = parallel_chain | prompt | llm | parser
```

```python
main_chain.invoke('HTTP')
```
